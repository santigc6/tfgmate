### In this notebook I will be using SAGE's division algorithm

```python
f(x,y) = 2*x + 3*y 
g(x,y) = 4*x^2 + 2*y^2
```

```python
q, r = g.maxima_methods().divide(f)
```

```python
print 'quotient: ', q
print 'remainder: ', r
```

```python
x,y,z = QQ['x,y,z'].gens()
I = ideal(x^5 + y^4 + z^3 - 1,  x^3 + y^3 + z^2 - 1)
B = I.groebner_basis()
print B
print len(B)
```

```python
f,g,h = B
(2*x*f + g).reduce(B)
```

```python
(2*x*f + g) in I
```

```python
(2*x*f + 2*z*h + y^3).reduce(B)
```

```python
(2*x*f + 2*z*h + y^3) in I
```
