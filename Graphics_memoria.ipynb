```python
P.<x,y> = PolynomialRing(QQ, 2, order='lex')

f=y^5 - 6*y^3 + 8 + x^4 + x^3
g=x*y - 2*x -5*y
I = ideal(f, g)
B = I.groebner_basis()
print B
print len(B)
```

```python
def is_GB_worth(GB): # Receives a GB
    n_depend1=0
    n_depend2=0
    for pol in GB:
        depend_x=False
        depend_y=False
        
        # Comprobamos si depende de x y de y
        if pol.degree(x) > 0:
            depend_x=True
        if pol.degree(y) > 0:
            depend_y=True
            
        # Actualizamos el numero de de polinomios que dependen de cada cosa
        if depend_x and depend_y:
            n_depend2 += 1
        elif depend_x or depend_y:
            n_depend1 += 1
    if n_depend1 >= n_depend2:
        return True
    else:
        return False

is_GB_worth(B)
```
